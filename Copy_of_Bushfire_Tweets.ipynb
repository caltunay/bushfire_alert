```
import pandas as pd 
import numpy as npaiohttp

import nest_asyncio
nest_asyncio.apply()

import twint
import pandas

import time
```

```
# extract tweets and create dataframe by list of keywords
# for australia
# post 2018
def scrape_keywords(keywords):
    df = pd.DataFrame() # empty df to append tweets for each keyword within loop

    # twint config  
    c = twint.Config()
    
    c.Lang = "en"
    c.Near = 'Australia'
    c.Since = '2018-01-01'
    c.Hide_output = True
    # c.Store_csv = True
    c.Pandas = True
    
    # assert input is list
    if type(keywords) != list:
        keywords = [keywords]

    # loop through keywords, extract tweets and append to df
    for keyword in keywords:
        
        c.Search = keyword
        # c.Output = f"{keyword}.csv"
        twint.run.Search(c)
        df_temp = twint.storage.panda.Tweets_df # tweets df for keyword

        try:
            if len(df) == 0 and len(df_temp) != 0: 
                df_temp['keyword'] = keyword # add current keyword to dataframe
                df = df_temp.copy()
            elif len(df) != 0 and len(df_temp) != 0:
                df_temp['keyword'] = keyword # add current keyword to dataframe
                df = df.append(df_temp, ignore_index = True)
        except:
            pass

    df.to_csv('tweets_df.csv') # output all tweets to csv
    return df 
```
